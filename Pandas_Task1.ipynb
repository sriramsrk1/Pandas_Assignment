# Q12. Create another DataFrame dept_df and merge with df horizontally. Store it in a variable called merged_df

dept_df columns:

| Name    | Department |
| ------- | ---------- |
| Alice   | HR         |
| Bob     | IT         |
| Charlie | Finance    |
| David   | Marketing  |
| Eva     | Admin      |




```python
# YOUR CODE HERE


dept_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Admin']
}
dept_df = pd.DataFrame(dept_data)


merged_df = pd.merge(df, dept_df, on='Name')
print(merged_df)
#raise NotImplementedError()
```

# Q13. Concatenate two DataFrames vertically.

```python
# YOUR CODE HERE
merged_df = pd.merge(df, dept_df, on='Name')
concatenated_df = pd.concat([df, merged_df], axis=0, ignore_index=True)

print(concatenated_df)

#raise NotImplementedError()
```

# Q14. Replace any missing Age values with the mean of the column.

```python
# YOUR CODE HERE
mean_age = concatenated_df['Age'].mean()
concatenated_df['Age'].fillna(value=mean_age, inplace=True)

print(concatenated_df)
#raise NotImplementedError()
```
